```python
import tkinter as tk
import time
from datetime import datetime


def update_clock():
    now = datetime.now()

    if is_24_hour.get():
        time_str = now.strftime("%H:%M:%S")
    else:
        time_str = now.strftime("%I:%M:%S %p")

    date_str = now.strftime("%A, %d %B %Y")

    time_label.config(text=time_str)
    date_label.config(text=date_str)

    
    root.after(1000, update_clock)


def toggle_format():
    if is_24_hour.get():
        format_button.config(text="Switch to 12-hour")
    else:
        format_button.config(text="Switch to 24-hour")

    update_clock()


root = tk.Tk()
root.title("Advanced Digital Clock")
root.geometry("400x250")
root.resizable(False, False)
root.configure(bg="#0F0F0F")


is_24_hour = tk.BooleanVar(value=True)


time_label = tk.Label(root,
                      font=("Segoe UI", 48, "bold"),
                      fg="#39FF14", bg="#0F0F0F")
time_label.pack(pady=(30, 0))


date_label = tk.Label(root,
                      font=("Segoe UI", 18),
                      fg="#FFFFFF", bg="#0F0F0F")
date_label.pack(pady=(10, 20))


format_button = tk.Checkbutton(root,
                               text="Switch to 12-hour",
                               font=("Segoe UI", 12),
                               variable=is_24_hour,
                               command=toggle_format,
                               fg="white", bg="#1A1A1A",
                               selectcolor="#1A1A1A",
                               activeforeground="white",
                               activebackground="#1A1A1A",
                               relief=tk.FLAT)
format_button.pack()


update_clock()


root.mainloop()
```
